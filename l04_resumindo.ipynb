# Resumindo
Nesta lição, você irá resumir textos com foco em tópicos específicos.

## Configuração

````python
from google import genai
from dotenv import load_dotenv
import os

# Carrega variáveis do .env
load_dotenv()
GOOGLE_API_KEY = os.getenv('API_KEY_IA202601')

client = genai.Client(api_key=GOOGLE_API_KEY)
````

````python
from google.genai import types

def get_completion(prompt):
    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=prompt,
        config=types.GenerateContentConfig(
            temperature=0,
            top_p=0.95,
            top_k=20,
        ),
    )
    return response.text
````

## Texto para resumir

````python
prod_review = """
Comprei este panda de pelúcia para o aniversário da minha filha, que adorou e leva para todo lugar. É macio e super fofo, e seu rosto tem uma expressão amigável. Achei um pouco pequeno pelo preço que paguei. Acho que pode haver outras opções maiores pelo mesmo valor. Chegou um dia antes do esperado, então pude brincar com ele antes de dar para ela.
"""
````

## Resuma com um limite de palavras/frases/caracteres

````python
prompt = f"""
Sua tarefa é gerar um resumo curto de uma avaliação de produto de um site de e-commerce.

Resuma a avaliação abaixo, delimitada por três crases, em no máximo 30 palavras.

Avaliação: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````

## Resuma com foco em envio e entrega

````python
prompt = f"""
Sua tarefa é gerar um resumo curto de uma avaliação de produto de um site de e-commerce para dar feedback ao departamento de Entregas.

Resuma a avaliação abaixo, delimitada por três crases, em no máximo 30 palavras, e foque em qualquer aspecto que mencione envio e entrega do produto.

Avaliação: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````

## Resuma com foco em preço e valor

````python
prompt = f"""
Sua tarefa é gerar um resumo curto de uma avaliação de produto de um site de e-commerce para dar feedback ao departamento de precificação, responsável por determinar o preço do produto.

Resuma a avaliação abaixo, delimitada por três crases, em no máximo 30 palavras, e foque em qualquer aspecto relevante ao preço e valor percebido.

Avaliação: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````

````python
for i in range(len(reviews)):
    prompt = f"""
    Sua tarefa é gerar um resumo curto de uma avaliação de produto de um site de e-commerce.

    Resuma a avaliação abaixo, delimitada por três crases, em no máximo 20 palavras.

    Avaliação: ```{reviews[i]}```
    """

    response = get_completion(prompt)
    print(i, response, "\n")
````
