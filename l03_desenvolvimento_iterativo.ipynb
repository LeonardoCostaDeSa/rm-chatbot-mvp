# Desenvolvimento Iterativo de Prompts
Nesta lição, você irá analisar e refinar iterativamente seus prompts para gerar textos de marketing a partir de uma ficha técnica de produto.

## Configuração

````python
from google import genai
from dotenv import load_dotenv
import os

# Carrega variáveis do .env
load_dotenv()
GOOGLE_API_KEY = os.getenv('API_KEY_IA202601')

client = genai.Client(api_key=GOOGLE_API_KEY)
````

````python
from google.genai import types

def get_completion(prompt):
    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=prompt,
        config=types.GenerateContentConfig(
            temperature=0,
            top_p=0.95,
            top_k=20,
        ),
    )
    return response.text
````

## Gere uma descrição de produto de marketing a partir de uma ficha técnica de produto

````python
fact_sheet_chair = """
VISÃO GERAL
- Parte de uma bela família de móveis de escritório inspirados no design de meados do século, incluindo arquivos, mesas, estantes, mesas de reunião e mais.
- Várias opções de cor da concha e acabamentos da base.
- Disponível com encosto e estofamento frontal em plástico (SWC-100) ou totalmente estofado (SWC-110) em 10 opções de tecido e 6 de couro.
- Opções de acabamento da base: aço inoxidável, preto fosco, branco brilhante ou cromo.
- Cadeira disponível com ou sem apoios de braço.
- Adequada para ambientes residenciais ou comerciais.
- Qualificada para uso em contratos.

CONSTRUÇÃO
- Base de alumínio revestida de plástico com 5 rodas.
- Ajuste pneumático da cadeira para fácil elevação/abaixamento.

DIMENSÕES
- LARGURA 53 CM | 20.87”
- PROFUNDIDADE 51 CM | 20.08”
- ALTURA 80 CM | 31.50”
- ALTURA DO ASSENTO 44 CM | 17.32”
- PROFUNDIDADE DO ASSENTO 41 CM | 16.14”

OPÇÕES
- Opções de rodízios para piso macio ou duro.
- Duas opções de densidade da espuma do assento: média (1.8 lb/ft3) ou alta (2.8 lb/ft3)
- Sem braços ou com apoios de braço PU de 8 posições

MATERIAIS
BASE DA CONCHA DESLIZANTE
- Alumínio fundido com revestimento de nylon PA6/PA66 modificado.
- Espessura da concha: 10 mm.
ASSENTO
- Espuma HD36

PAÍS DE ORIGEM
- Itália
"""
````

````python
prompt = f"""
Sua tarefa é ajudar uma equipe de marketing a criar uma descrição para o site de varejo de um produto com base em uma ficha técnica.

Escreva uma descrição do produto com base nas informações fornecidas nas especificações técnicas delimitadas por três crases.

Especificações técnicas: ```{fact_sheet_chair}```
"""
response = get_completion(prompt)
print(response)
````
