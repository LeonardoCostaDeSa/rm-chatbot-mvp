# Transformando

Neste notebook, vamos explorar como usar Modelos de Linguagem de Grande Escala para tarefas de transformação de texto, como tradução de idiomas, verificação ortográfica e gramatical, ajuste de tom e conversão de formato.

## Configuração

````python
from google import genai
from dotenv import load_dotenv
import os

# Carrega variáveis do .env
load_dotenv()
GOOGLE_API_KEY = os.getenv('API_KEY_IA202601')

client = genai.Client(api_key=GOOGLE_API_KEY)
````

````python
from google.genai import types

def get_completion(prompt):
    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=prompt,
        config=types.GenerateContentConfig(
            temperature=0,
            top_p=0.95,
            top_k=20,
        ),
    )
    return response.text
````

## Tradução

O ChatGPT é treinado com fontes em muitos idiomas. Isso dá ao modelo a capacidade de traduzir. Aqui estão alguns exemplos de como usar essa capacidade.

````python
prompt = f"""
Traduza o seguinte texto em inglês para espanhol: \ 
```Hi, I would like to order a blender```
"""
response = get_completion(prompt)
print(response)
````

````python
prompt = f"""
Diga-me qual idioma é este: 
```Combien coûte le lampadaire?```
"""
response = get_completion(prompt)
print(response)
````

````python
prompt = f"""
Traduza o seguinte texto para francês, espanhol
e inglês pirata: \
```I want to order a basketball```
"""
response = get_completion(prompt)
print(response)
````

````python
prompt = f"""
Traduza o seguinte texto para espanhol nas formas \
formal e informal: 
```Would you like to order a pillow?```
"""
response = get_completion(prompt)
print(response)
````

### Tradutor Universal
Imagine que você é responsável de TI em uma grande empresa multinacional de e-commerce. Usuários estão te enviando mensagens com problemas de TI em todos os seus idiomas nativos. Sua equipe é de todo o mundo e fala apenas seus idiomas nativos. Você precisa de um tradutor universal!

````python
user_messages = [
  "La performance du système est plus lente que d'habitude.",  # O desempenho do sistema está mais lento que o normal         
  "Mi monitor tiene píxeles que no se iluminan.",              # Meu monitor tem pixels que não acendem
  "Il mio mouse non funziona",                                 # Meu mouse não está funcionando
  "Mój klawisz Ctrl jest zepsuty",                             # Minha tecla Ctrl está quebrada
  "我的屏幕在闪烁"                                               # Minha tela está piscando
] 
````

````python
for issue in user_messages:
    prompt = f"Diga-me qual idioma é este: ```{issue}```"
    lang = get_completion(prompt)
    print(f"Mensagem original ({lang}): {issue}")

    prompt = f"""
    Traduza o seguinte texto para inglês
e coreano: ```{issue}```
    """
    response = get_completion(prompt)
    print(response, "\n")
````

## Conversão de Formato
O ChatGPT pode traduzir entre formatos. O prompt deve descrever os formatos de entrada e saída.

## Verificação Ortográfica/Gramatical.

Aqui estão alguns exemplos de problemas comuns de gramática e ortografia e a resposta do LLM.

Para sinalizar ao LLM que você quer que ele revise seu texto, instrua o modelo a 'revisar' ou 'revisar e corrigir'.

````python
text = [ 
  "The girl with the black and white puppies have a ball.",  # A menina com os filhotes preto e branco tem uma bola.
  "Yolanda has her notebook.", # ok
  "Its going to be a long day. Does the car need it’s oil changed?",  # Homônimos
  "Their goes my freedom. There going to bring they’re suitcases.",  # Homônimos
  "Your going to need you’re notebook.",  # Homônimos
  "That medicine effects my ability to sleep. Have you heard of the butterfly affect?", # Homônimos
  "This phrase is to cherck chatGPT for speling abilitty"  # ortografia
]
for t in text:
    prompt = f"""Revisar e corrigir o texto a seguir
e reescrever a versão corrigida. Se não encontrar
erros, apenas diga "Nenhum erro encontrado". Não use
qualquer pontuação ao redor do texto:
```{t}```"""
    response = get_completion(prompt)
    print(response)
````

````python
text = f"""
Comprei isto para minha filha no aniversário dela porque ela sempre pega o meu do meu quarto. Sim, adultos também gostam de pandas. Ela leva para todo lugar, e é super macio e fofo. Uma das orelhas é um pouco mais baixa que a outra, e não acho que isso foi feito para ser assimétrico. Achei um pouco pequeno pelo preço que paguei. Acho que pode haver outras opções maiores pelo mesmo valor. Chegou um dia antes do esperado, então pude brincar com ele antes de dar para minha filha.
"""
prompt = f"revisar e corrigir esta avaliação: ```{text}```"
response = get_completion(prompt)
print(response)
````

````python
from redlines import Redlines

diff = Redlines(text,response)
display(Markdown(diff.output_markdown))
````

````python
prompt = f"""
revisar e corrigir esta avaliação. Torne-a mais envolvente. 
Garanta que siga o guia de estilo APA e seja direcionada a um leitor avançado. 
Saída em formato markdown.
Texto: ```{text}```
"""
response = get_completion(prompt)
display(Markdown(response))
````
