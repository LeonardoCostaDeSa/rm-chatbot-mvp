## Configuração

````python
from google import genai
from dotenv import load_dotenv
import os

# Carrega variáveis do .env
load_dotenv()
GOOGLE_API_KEY = os.getenv('API_KEY_IA202601')

client = genai.Client(api_key=GOOGLE_API_KEY)
````

````python
from google.genai import types

def get_completion(prompt):
    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=prompt,
        config=types.GenerateContentConfig(
            temperature=0,
            top_p=0.95,
            top_k=20,
        ),
    )
    return response.text
````

## Personalize a resposta automática para um e-mail de cliente

````python
# dado o sentimento da lição sobre "inferência",
# e a mensagem original do cliente, personalize o e-mail
sentiment = "negativo"

# avaliação para um liquidificador
review = f"""
Ainda tinham o sistema de 17 peças em promoção sazonal por cerca de R$ 49 em novembro, quase metade do preço, mas por algum motivo (chame de aumento de preço) na segunda semana de dezembro os preços subiram para cerca de R$ 70-89 pelo mesmo sistema. E o sistema de 11 peças subiu cerca de R$ 10 também em relação ao preço promocional anterior de R$ 29. Parece bom, mas se você olhar para a base, a parte onde a lâmina encaixa não parece tão boa quanto em edições anteriores de alguns anos atrás, mas pretendo ser bem cuidadoso (por exemplo, trituro itens duros como feijão, gelo, arroz, etc. no liquidificador primeiro, depois pulverizo na quantidade que quero e troco para a lâmina de bater para uma farinha mais fina, e uso a lâmina cruzada primeiro para fazer smoothies, depois uso a lâmina reta se quiser mais fino/menos polposo). Dica especial para smoothies: corte e congele as frutas e vegetais (se usar espinafre, refogue levemente para amolecer e congele até usar - e se for fazer sorvete, use um processador de alimentos pequeno ou médio) que você pretende usar, assim evita adicionar tanto gelo ao fazer seu smoothie. Depois de cerca de um ano, o motor começou a fazer um barulho estranho. Liguei para o atendimento ao cliente, mas a garantia já tinha acabado, então tive que comprar outro. Observação: a qualidade geral desses produtos caiu, então eles estão contando com o reconhecimento da marca e a fidelidade do consumidor para manter as vendas. Chegou em cerca de dois dias.
"""
````

````python
prompt = f"""
Você é um assistente de IA de atendimento ao cliente.
Sua tarefa é enviar uma resposta por e-mail a um cliente valioso.
Dado o e-mail do cliente delimitado por ```, gere uma resposta agradecendo pela avaliação.
Se o sentimento for positivo ou neutro, agradeça pela avaliação.
Se o sentimento for negativo, peça desculpas e sugira que ele entre em contato com o atendimento ao cliente.
Certifique-se de usar detalhes específicos da avaliação.
Escreva de forma concisa e profissional.
Assine o e-mail como `Agente de atendimento IA`.
Avaliação do cliente: ```{review}```
Sentimento da avaliação: {sentiment}
"""
response = get_completion(prompt)
print(response)
````
